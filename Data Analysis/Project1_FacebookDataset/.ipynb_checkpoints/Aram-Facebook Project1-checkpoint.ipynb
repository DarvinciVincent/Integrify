**Research Question 1: Which advertisers have the highest number of political ads?**

```python
# Research Question 1: Which advertisers have the highest number of political ads?
top_advertisers = df[df['political'] > 0]['advertiser'].value_counts().nlargest(5)
print("Top Advertisers with Highest Number of Political Ads:")
print(top_advertisers)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_advertisers.index, y=top_advertisers.values)
plt.title('Top Advertisers with Highest Number of Political Ads')
plt.xlabel('Advertiser')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Research Question 2: Are there specific entities or entity-type that appear frequently in political ads?**

```python
entity_counter = Counter()
entity_type_counter = Counter()

political_ads_df = df[df['political'] > 0]

for row in political_ads_df['entities']:
    for dictionary in row:
        entity = dictionary['entity']
        entity_type = dictionary['entity_type']
        entity_counter[entity] += 1
        entity_type_counter[entity_type] += 1

# Print the top 10 most frequent entities in political ads
print('Top 10 most frequent entities in political ads:')
for entity, count in entity_counter.most_common(10):
    print(entity, '-', count)

# Print the top 10 most frequent entity types in political ads
print('\nTop 10 most frequent entity types in political ads:')
for entity_type, count in entity_type_counter.most_common(10):
    print(entity_type, '-', count)
```

```python
entity_pair_counter = Counter()
political_ads_df = df[df['political'] > 0]

for row in political_ads_df['entities']:
    entities = row
    entity_pairs = [(entity['entity'], entity['entity_type']) for entity in entities]
    entity_pair_counter.update(entity_pairs)

# Get the top 10 most frequent entity-entity_type pairs and their counts
top_entity_pairs = entity_pair_counter.most_common(10)
top_entity_pair_labels = [f"{pair[0][0]} ({pair[0][1]})" for pair in top_entity_pairs]
top_entity_pair_counts = [pair[1] for pair in top_entity_pairs]

# Plot the bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_entity_pair_labels, top_entity_pair_counts, color='blue')
plt.title('Top Entity-Entity_Type Pairs in Political Ads')
plt.xlabel('Entity-Entity_Type Pair')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
