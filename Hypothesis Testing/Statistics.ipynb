- ### 1.3. Measure of center: 

    It is a value at a center/middle of data set (a number that is in some way ***central*** to the set).

    - ***Mean/Average***: also known as expected value in math (sum over the number of data). 
    - ***Median***: represent the element in the middle of a sorted data.

```python
import numpy as np
import statistics as st

values = np.array([1, 1, 2, 2, 2, 3, 4, 5, 3, 3, 7, 7, 9])
 
print('Mean/average =', round(np.mean(values), 2))
print('Median =', np.median(values))
print(f"The Mode is {st.mode(values)} (most frequently occurring number)")
print('Mid-range =', (values.max() + values.min())/2)
values.shape, values.size
```

- ### 1.4. Measure of spread: shows how our data set is spread.

  1. ***Quartiles***: are values that separate the data into four equal parts ***(Q0, Q1, Q2, Q3, Q4)***.
     - ***Min Q0***: is the smallest value in the data set. 
     - ***First Quartile Q1***: The value such that ***25% percentile*** of the data falls below.
     - ***Second Quartile Q2(≈ median)***: The value such that ***50% percentile*** of the data falls below
     - ***Third Quartile Q3***: The value such that ***75% percentile***  of the data falls below.
     - ***Max Q4***: is the largest value in the data set.
  2. ***Percentiles***: are values that separate the data into 100 equal parts.
     - ***The 25th percentile (P25%)*** is the same as the first quartile (Q1).
     - ***The 50th percentile (P50%)*** is the same as the second quartile (Q2) and the median.
     - ***The 75th percentile (P75%)*** is the same as the third quartile (Q3)                
  3. ***Range***: is the difference between the maximum and the minimum.
  4. ***Interquartile Range (IQR)***: is calculated as the difference between Q3​ and Q1​.
  5. ***Variance***: is used to compare the spread of two different groups.
  $$
  \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}
  $$
  6. ***Standard deviation***: is used more often in practice than the variance because it shares the units of the original dataset.
$$
  \sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}}
$$

```python
import numpy as np
import statistics as st

values = [13,21,21,40,42,48,55,72]

print('Min =', np.min(values), '(Q0)=',np.quantile(values, 0))
print('First Quartile (Q1)=', np.quantile(values, 0.25))
print('Second Quartile (Q2)/(≈ median)', np.median(values), np.quantile(values, 0.5))
print('Third Quartile (Q3)', np.quantile(values, 0.75))
print('Max (Q4)=', np.max(values), np.quantile(values, 1.))
```

## 2. Inferential statistics

### 2.1.Sampling Techniques

   - Why Sampling?
        - Big dataset (run out of memory).
        - less time than using the whole population
        - less cumbersome and more practical than an analysis of the entire population.
   - ***Population:*** represents the whole dataset.
   - ***Sample:*** is a subset of the population.
   - ***Observation/individual:*** is the study subject.

   - ***Sampling*** is a method that selects a subset to gather information about the entire population without the need to study its’ individual separately.

- Some aspects to consider prior to collecting a data sample include:
    - ***Population:*** The scope or domain from which observations could theoretically be made.
    - ***Sample Goal:*** The population property that you wish to estimate using the sample.
    - ***Selection Criteria:*** The methodology that will be used to accept or reject observations in your sample.
    - ***Sample Size:*** is the number of observations that will constitute the sample.

- There are two main categories for sampling techniques:

    1. **Probability Sampling**
    2. **Non-Probability Sampling**

#### 2.1.1. Probability Sampling:
   - Every element of the population has an equal chance of being selected. 
   - It gives us the best chance to create a sample that is truly representative of the population.

   - There four main probability sampling techniques:
       1. **Simple Random Sampling**
       2. **Systematic Sampling**
       3. **Stratified Sampling**
       4. **Clustering Sampling**
   
##### 2.1.1.1. Simple Random Sampling

- Every individual is chosen entirely by chance and each member of the population has an equal chance of being selected.
- It is the most direct method of probability sampling. 
- Reduces selection bias.
- It ***may not select enough individuals*** with our characteristics of interest.


```python
import random

names = ["Ali", "Jhon", "Tom", "Stella", "Aron", "Sofie", "Beth"]
print("The random sample:", random.sample(names, 4))
```

- Every individual of the population is randomly chosen with equal chance of being selected.

    - Suppose, we began with 3rd person, and we want a sample size of 5. 
    - The next individual person would be at an interval of (20/5) = 4 from the 3rd person, i.e. 7 (3+4),  and so on.

##### 2.1.1.3. Stratified Sampling

- We divide the population into subgroups (called strata) based on different traits like gender, category, etc. And then we select the sample(s) from these subgroups.

- We first divided our population into subgroups based on different colors of red, yellow, green and blue. 
- Then, from each color, we selected an individual in the proportion of their numbers in the population.

- We use this type of sampling when we want representation from all the subgroups of the population.
- Stratified sampling requires ***proper knowledge of the characteristics*** of the population.


- Example:

- Suppose we want to estimate the average salary of employees in a small company that has 21 employees. The company has five departments, and we want to ensure that our sample is representative of each department.

- To use stratified sampling, we first need to divide the population into strata, or subgroups, based on the characteristic of interest. In this case, we will divide the employees into five strata based on their department. Let's assume that the population is as follows:

- Department A: 4 employees, with salaries of 30,000, 35,000, 40,000, and 45,000
- Department B: 5 employees, with salaries of 25,000, 30,000, 35,000, 40,000, and 45,000
- Department C: 3 employees, with salaries of 40,000, 45,000, and 50,000
- Department D: 4 employees, with salaries of 35,000, 40,000, 45,000, and 50,000
- Department E: 4 employees, with salaries of 30,000, 35,000, 40,000, 45,000, and 45,000

##### 2.1.1.4. Clustering Sampling

- We use the subgroups of the population as the sampling unit rather than individuals. 
- The population is divided into subgroups, known as clusters, and a whole cluster is randomly selected to be included in the study.

- We have divided our population into 5 clusters. Each cluster consists of 4 individuals and we have taken the 4th cluster in our sample. We can include more clusters as per our sample size.

- This type of sampling is used when we ***focus on a specific region or area***.

#### 2.1.2. Non-Probability Sampling:
   - All elements do not have an equal chance of being selected.
   - There is a significant risk of ending up with a non-representative sample which does not produce generalizable results.
   
- There four main Non-probability sampling techniques:

    1. Convenience Sampling
    2. Quota Sampling
    3. Judgment Sampling
    4. Snowball sampling


##### 2.1.2.1. Convenience Sampling:
- Very easy method because individuals are selected based on their availability and willingness to take part.

- Here, let’s say individuals numbered 4, 7, 12, 15 and 20 want to be part of our sample, and hence, we will include them in the sample.

- Convenience sampling is ***prone to significant bias***, because the sample ***may not be the representation*** of the specific characteristics such as religion or, say the gender, of the population.


##### 2.1.2.2. Quota Sampling

- We choose items based on predetermined characteristics of the population.

- ***Example:***
    - Lets have a popilation of 20 person. 
    - Considering that we have to select individuals having a number in multiples of four for our sample. Therefore, the individuals numbered 4, 8, 12, 16, and 20 are already reserved for our sample.


- In quota sampling, the chosen sample ***might not be the best representation*** of the characteristics of the population that weren’t considered


##### 2.1.2.3. Judgment Sampling

- known as ***selective sampling***. It depends on the judgment of the experts when choosing whom to ask to participate.

- ***Example:*** 
    - Lets have a popilation of 20 person. 
    - Suppose, our experts believe that people numbered 1, 7, 10, 15, and 19 should be considered for our sample as they may help us to infer the population in a better way. 

- As you can imagine, quota sampling is also ***prone to bias*** by the experts and ***may not necessarily be representative-***.


##### 2.1.2.4. Snowball sampling

- Existing people are asked to nominate further people known to them so that the sample increases in size like a rolling snowball. 
- This method of sampling is effective when a sampling frame is difficult to identify.

- ***Example:***
    - Lets have a popilation of 20 person. 
    - Here, we had randomly chosen person 1 for our sample, and then he/she recommended person 6, and person 6 recommended person 11, and so on.

- There is a ***significant risk of selection bias*** in snowball sampling, as the referenced individuals will share common traits with the person who recommends them.

#### 3.1.1. Confidence Interval:

- ***Confidence Interval (CI):*** The bounds (range of values) on the estimate of a population parameter (meanm, variance, or standard deviation) lies with certain probability.
    - CI requires two things:
        - Point of estimate (sample mean).
        - Interval estimate.
        
- To estimate the confidence interval of a normaly distributed population $\bar x$:

 $$
  CI = \bar x + z \times \frac{\sigma} {\sqrt{n}}
  $$
  
$$
z = 
\begin{cases}
1.96 \text{ if } Confidence = 95\% \\
2.58 \text{ if } Confidence = 99\%
\end{cases}
$$


- ***Example:*** if the 95% CI of the average fill volume of 375 ml bottles is 368–372 ml, you can be 95% confident that the true value of the process mean is within this interval.


```python
import random
import numpy as np
import scipy.stats as st

#define sample data
population = [12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29] # generaly se dont have the whole population
sample = random.sample(population, 6)

# create 95% confidence interval for population mean weight

smean = np.mean(sample)
s = np.std(sample)

LCL = smean - 1.96*(s/np.sqrt(len(sample)))
UCL = smean + 1.96*(s/np.sqrt(len(sample)))

print(sample)
print('The population mean %f with 95 confidence lies between [%f, %f]'%(np.array(population).mean(), LCL,UCL))
# st.t.interval(alpha=0.95, df=len(data)-1, loc=np.mean(data), scale=st.sem(data))
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate some random data
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.randn(100)

# Compute the sample mean and standard deviation
x_mean = np.mean(x)
y_mean = np.mean(y)
y_std = np.std(y)

# Set the confidence level and compute the critical value
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = np.abs(norm.ppf(alpha / 2))

# Compute the margin of error
margin_of_error = z_critical * y_std / np.sqrt(len(y))

# Plot the data points and the confidence interval region
plt.scatter(x, y, s=10, color='blue')
plt.plot(x, 2 * x, color='green', label='True Line')
plt.fill_between(x, 2 * x - margin_of_error, 2 * x + margin_of_error, color='gray', alpha=0.2, label='95% Confidence Interval')
plt.legend(loc='upper left')
plt.show()
```

### 3.2. Hypothesis Testing

- ***What is a Hypothesis?***
    - A hypothesis is an guess about something in the world around you either by experiment or observation.
    - **Example**: 
        - A new medicine you think might work.
        - A way of teaching you think might be better.
        - A possible location of new species.

- ***What is a Hypothesis Statement?***
    - Hypothesis requires writing an explicit statement
    - If I…(do this to an independent variable)….then (this will happen to the dependent variable).

        - **Example**: 
            - If I (decrease the amount of water given to herbs) then (the herbs will increase in size).
            
- ***What is a Hypothesis Testing?***
    - In statistics is a way to test the results of a survey or experiment to see if the results are meaningful and valid. 
    - Basically testing whether the results are valid by figuring out the odds that have happened by chance. 
    - If the results may have happened by chance, the experiment won’t be repeatable and so has little use.

    - What need to know is:
        - Figure out the null hypothesis, also alternative hypothesis
        - State the null hypothesis,
        - Choose what kind of test need to perform,
        - Either accept or reject the null hypothesis.

- ***What is a Null Hypothesis?***
    - In history of science, the null hypothesis H0 is always the accepted fact.
    - Opposite of alternative hypothesis H1
    - Researchers work to reject, nullify or disprove the H0.
        - Not including the null hypothesis in your research is considered very bad practice by the scientific community.
    - **Examples:**
        - **Null hypothesis H0**: The world is flat.
        - **Alternate hypothesis H1**: The world is round.

- ***How do we State the Null Hypothesis?***
    - In practice you will be given a problem from real-life situations. 
    - You’ll need to figure out what your hypothesis is from the problem. 
    - you are looking to find a fact that is nullifiable (i.e. something you can reject).
    
        - **Example**: A researcher thinks that if knee surgery patients go to physical therapy twice a week (instead of 3 times), their recovery period will be longer. Average recovery times for knee surgery patients is 8.2 weeks.
            - **Hypothesis Statement**: 
                - H0: the researcher believes the average recovery time is less or equal than 8.2 weeks (H0: 𝜇 ≤ 8.2)
                - H1: the researcher believes the average recovery time is more than 8.2 weeks (H1: 𝜇 > 8.2)

- To conduct a one-sample hypothesis test, we need to follow these steps:
    - Define the null and alternative hypotheses.
    - Set the significance level.
    - Calculate the test statistic.
    - Determine the p-value.
    - Make a decision and draw a conclusion.

    - The null hypothesis (H0) is that the population mean height of the trees is 35 inches.

        - H0: μ = 35
        - Ha: μ ≠ 35
        - where μ is the population mean height.
- Step 2: Set the significance level.
    - Let's set the significance level at α = 0.05.
- Step 3: Calculate the test statistic.
    - To calculate the test statistic, we can use the t-distribution since the sample size is small (n < 30) and the population standard deviation is unknown.
    - The test statistic is calculated as:
    - t = (x̄ - μ) / (s / sqrt(n))
    - where x̄ is the sample mean, μ is the hypothesized population mean, s is the sample standard deviation, and n is the sample size.
- Step 4: Determine the p-value.
    - To determine the p-value, we need to find the probability of observing a test statistic as extreme or more extreme than the one we calculated, assuming the null hypothesis is true.
    - Since this is a two-tailed test, we need to find the area in both tails of the t-distribution with n - 1 degrees of freedom, where n is the sample size.

#### 4.2.2. Compare difference in sample (T- test) 
- T-test is a statistical test that is used to compare the unknown population(s) means of two groups (N $\leq$ 30) are equal or not.
$$
  t = \frac{\bar{x} - \mu}{\frac{S}{\sqrt{n}}}
$$

- ***When to use a t-test ?***
    - A t-test can only be used when comparing the means of two groups . If you want to compare more than two groups use an ANOVA test or a post-hoc test.

1. ***A parametric t-test assuming:***
    - Data are independent
    - Data are (approximately) normally distributed
    - Have a similar amount of variance within each group being compared 

2. ***Non-parametric t-test assuming:***
    - If your data do not fit these assumptions, you can try a nonparametric alternative to the t test, such as the Wilcoxon Signed-Rank test for data with unequal variances.

- ***One-sample, two-sample, or paired t-test?***
    - ***One-sample t-test:*** If there is one group being compared against a standard value (e.g., comparing the acidity of a liquid to a neutral PH = 7)
    - ***Paired t-test:*** If the groups come from a single population (e.g., measuring before and after an experimental treatment). This is a ***within-subjects design***.
    - ***Two-sample/independent t-test:*** If the groups come from two different populations (e.g., two different species, or people from two separate cities). This is a ***between-subjects design***.

- ***One-tailed or two-tailed t-test?***
    - If you only care whether the two populations are different from one another, perform a ***two-tailed t-test***.
    - If you want to know whether one population mean is greater than or less than the other, perform a ***one-tailed t-test***.

#### 4.2.4. hypothesis testing (Anova test)

- ***ANOVA-test*** is a statistical test that is used Analysis of variance. ANOVA assesses the differences between group means of at least two populations are different. 
    - Continuous dependent variable and a categorical independent variable that divides your data into comparison groups to perform ANOVA.
    - <strong style="color: red; opacity: 0.80;">if the ANOVA is significant one cannot tell which group is different.</strong>

- ***One-way/One-Factor ANOVA-test***: is a parametric test used to test for a statistically significant difference of an outcome between 3 or more groups (i.e. at least one of the groups is statistically significantly different than the others).
    - One-Factor refers to the number of variables in the analysis and not to the groups.
    - The independent variables are refers as factors
    - The number of groups within each variable are called levels ( i.e. one variable with 3 categories could be referred to as a factor with 3 levels)


- **Parametric test assumptions**
    - Population distributions are normal
    - Samples have equal variances
    - Independence
- **Hypothesis**
    - $H_{0}$ = 𝜇1 = 𝜇2 =𝜇3 =𝜇4 =𝜇... =𝜇k
    - $H_{0}$ = at least one of the group means differs.
    - The test statistic is the **F-statistic** and compares the mean square between samples **($MS_{Between}$)** to the mean square within sample **($MS_{Within}$)**.
    - and where,
        - **$SS_{B}$** is the sum um of square between sample 
        - **$SS_{W}$** is  the sum of square within sample

        - **$n_{T}$** is the total number of observations
        - **k** is the number of groups
        
$$
  F-statistic = \frac{MS_{Between}}{MS_{Within}}
$$ 
$$ 
  MS_{Between} = \frac{SS_{Between}}{df_{Between}}, SS_{Between} = \sum n_{k} (\bar x_{k} - \bar x)^{2}, df_{Between} = k-1
$$ 
$$ 
  MS_{Within} = \frac{SS_{Within}}{df_{Within}}, SS_{Within} = \sum n_{k} (\bar x_{i, k} - \bar x)^{2}, or SS_{Within} = \sum (\bar n_{k} - 1)^{2}, df_{Within} = n_{k}-1
$$        

- The F-statistic measures the ratio of the between-group variability to the within-group variability. 
- A larger F-statistic indicates a larger difference between the group means relative to the variability within each group, and thus a stronger evidence against the null hypothesis of equal group means. 
- The p-value represents the probability of observing an F-statistic as large or larger than the one obtained under the null hypothesis, assuming the null hypothesis is true. 
- A small p-value (e.g., less than 0.05) indicates strong evidence against the null hypothesis and suggests that at least one group mean is significantly different from the others.

```python
import numpy as np
from scipy.stats import f, f_oneway

# 𝐻0 = 𝜇1 = 𝜇2 = 𝜇3
# 𝐻0 = at least one of the group means differs.
    
group1 = [3, 2, 1, 1, 4]
group2 = [5, 2, 4, 2, 3]
group3 = [7, 4, 5, 3, 6]

k = 3
n_T = 15

SS_B  = 5*pow((2.2 - 3.5),2) + 5*pow((3.2 - 3.5),2) + 5*pow((5.0 - 3.5),2)
df_B = k - 1
MS_B  = SS_B /df_B

SS_W  = (5 - 1)*1.7 + (5 - 1)*1.7 + (5 - 1)*2.5
df_W = n_T - k
MS_W  = SS_W /df_W

TSS = SS_B + SS_B

F_statistic = MS_B / MS_W

p_value = 1 - f.cdf(F_statistic, df_B, df_W)

print('ANOVA using the formula')
print("MS B:", MS_B)
print("MS W:", MS_W)
print("F-statistic:", F_statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print('Reject the null hypothesis H0, There is one group mean different from the others')
else:
    print('Fail to reject the null hypothesis, all the group means are equal')

print('ANOVA using the library')
F_statistic, p_value = f_oneway(group1, group2, group3)
print("F-statistic:", F_statistic)
print("p-value:", p_value)
if p_value < 0.05:
    print('Reject the null hypothesis H0, There is one group mean different from the others')
else:
    print('Fail to reject the null hypothesis, all the group means are equal')
print(np.array(group1).mean(), np.array(group2).mean(), np.array(group3).mean())
```

- ***Two-way/Two-Factors ANOVA-test:*** is used to determine whether or not there is a statistically significant difference between the means of three or more independent groups that have been split on two factors.

- The purpose of a two-way ANOVA is to determine **how two factors impact a response variable**, and to determine **whether or not there is an interaction between the two factors on the response variable**.

**Example:** A botanist wants to know whether or not plant growth is influenced by sunlight exposure and watering frequency. She plants 30 seeds and lets them grow for two months under different conditions for sunlight exposure and watering frequency. After two months, she records the height of each plant, in inches.

- We want to examine the relationship between the response variable ***height*** and two categorical predictor variables
    ***water*** and ***sun***, as well as the ***interaction*** between the predictor variables.
https://naneja.github.io/python/anova

```python
import numpy as np
import pandas as pd

# f(x1, x2) = y

# x1: watering (Categorical)
# x2: sun exposure (Categorical)
# y: plant growth (Numerical)

#create data
df = pd.DataFrame({'water': np.repeat(['daily', 'weekly'], 15),
                   'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
                   'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5,
                              6, 6, 7, 8, 7, 3, 4, 4, 4, 5,
                              4, 4, 4, 4, 4, 5, 6, 6, 7, 8]})

df#[:10]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ols: ordinary least square

# We want to examine the relationship between the response variable 'height' and two categorical predictor variables
# 'water' and 'sun', as well as their interaction term 'water:sun'.

#perform two-way ANOVA
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()#

# C() is used to indicate that the variables are categorical

# 'C(water):C(sun)' is the interaction term between the two categorical variables, 
# This allows us to examine whether the effect of water on height depends on the level of sun and vice versa.

sm.stats.anova_lm(model, typ=2)
```

- **Interpret the results**:
- We can see the following p-values for each of the factors in the table:
    - water: p-value = .000527
    - sun: p-value = .0000002
    - water*sun: p-value = .120667
- Since the p-values for water and sun are both less than .05, this means that both factors have a statistically significant effect on plant height.
- Since the p-value for the interaction effect (.120667) is not less than .05, this tells us that there is no significant interaction effect between sunlight exposure and watering frequency.
