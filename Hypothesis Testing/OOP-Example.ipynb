```python
default = "2022/10/20 13:31:26"
new_year = "2022/01/01 00:00:01"
delta = timedelta(hours=52, minutes=34, seconds=12)

str_default = datetime.strptime(default, "%Y/%m/%d %H:%M:%S")
day_default = str_default.strftime("%A")
month_default = str_default.strftime("%B")

str_NY = datetime.strptime(new_year, "%Y/%m/%d %H:%M:%S")
print_NY = str_NY.strftime("%A")

diff = str_default - str_NY

print(f"The month is {month_default}.")
print(f"The day is {day_default}.")
print(f"The New Year day was {print_NY}.")
print(f"It has been {diff.days} days since New Year.")
```

```python
# abstract class
class part:
    def __init__(self, shape, size):
        self.shape = shape
        self.size = size
    def get_part(self):
        print(self.shape, self.size)
    def set_shape(self, shape):
        self.shape = shape
    def set_size(self, size):
        self.size = size
        
# abstract class
class eye(part):
    def __init__(self, shape, size, color):
        super().__init__(shape, size)
        self.color = color
    def get_eye(self):
        self.get_part()
        print(self.color)
    def set_color(self, color):
        self.color = color
        
# you can Interact oonly with this        
class face:
    def __init__(self, eye_shape, eye_size, eye_color, nose_shape, nose_size, ear_shape, ear_size):
        self.Eye = eye(eye_shape, eye_size, eye_color)
        self.ear = part(ear_shape, ear_size)
        self.nose = part(nose_shape, nose_size)
    def get_face(self):
        self.Eye.get_eye()
        self.ear.get_part()
        self.nose.get_part()

f = face('round', 'big', 'blue', 'paron', 'big', 'nut', 'small')
f.get_face()
```
